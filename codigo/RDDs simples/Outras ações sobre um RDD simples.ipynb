```python
# Quando chamado em um DStream de elementos do tipo K, 
# devolve um novo DStream de pares (K, Long) onde o valor 
# de cada chave é sua freqüência em cada RDD da fonte DStream.

# ou seja devolve a quantidade que cada elemento aparece

rddCountBy = sc.parallelize(list("abracadabra")).cache()
mimapa = rddCountBy.countByValue()

print(type(mimapa))
print(mimapa.items())
```

```python
# top() devolve uma lista com os n primeiros elementos do RDD ordenado em ordem decrecente
rdd = sc.parallelize([8,4,2,9,3,1,10,5,6,7]).cache()
```

```python
print("4 maiores elementos : {0}".format(rdd.top(4)))
```

```python
print("4 menores elementos : {0}".format(rdd.takeOrdered(4)))
```

```python
# pega os menores elementos e devolve o oposto de cada um deles
print("4 maiores elementos : {0}".format(rdd.takeOrdered(4,lambda x : -x)))
```
